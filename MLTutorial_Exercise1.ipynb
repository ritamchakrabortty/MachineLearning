```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
dg = pd.read_csv("/content/canada_per_capita_income.csv")
dg
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income(US$)')
plt.scatter(dg.year,dg.per_capita_income, color='red',marker='*')
```

```python
reg = linear_model.LinearRegression()
reg.fit(dg[['year']],dg.per_capita_income)
```

```python
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per_capita_income(US$)')
plt.scatter(dg.year,dg.per_capita_income, color='red',marker='*')
plt.plot(dg.year,reg.predict(dg[['year']]),color='blue')
```
